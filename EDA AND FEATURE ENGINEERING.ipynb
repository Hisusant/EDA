# EDA - Exploratory Data Analysis

    It is a data exploration technique to understand various aspects of data. 

    Data science pipeline :

    1. Data collection/injestion
    2. EDA(analysis)
    3. Feature Engineering or Data Cleansing or preprocessing
    4. Model building
    5. Evaluation matrix or validation

## 2. EDA Steps :

    1. Profile of data
            1. shape of data(rows and columns)
            2. missing data
            3. catagorical data
            4. numerical data
            5. duplicates data
            6. data types 
    2. statistical interpretation
            1. variance
            2. covariance
            3. standard deviation
            4. corelation
            5. t-test
            6. z-test
            7. Annova test
            8. chi square test
            9. mean,median,mode
    3. Graph based analysis
            1. box plot
            2. scatter plot
            3. histogram
            4. countplot
            5. pie plot
            6. heatmap
            7. violine plot
            8. KDE
            9. pair plot
    
## 3. Feature Engineering or Data cleansing or Data preprocessing
    
    1. Missing value treatment
            1. random value implementation
            2. forward filling / backward filling
            3. statistical approach-mean/median/mode
            4. drop rows/columns
            5. KNN imputer
            6. ML algorithm
            7. ML model to predict missing value 
    2. Outlier treatment
        Detect outliers :
            1. Z-score
            2. IQR - Inter quartile range(Q3-Q1)
            3. Box plot
            4. Violine plot
            5. scatter plot
         Handling outliers :
             1. drop rows/columns
             2. median
             3. data trimming
    3. Scaling of data
            1. standardization 
            2. Normalization
            3. min-max scaller
            4. unit scalling
    4. Transformation 
            1. log transformation
            2. box-cos transformation
            3. square transformation
            4. cube transformation
            5. power transformation
    5. Encoding
            1. one hot encoding
            2. label encoding
            3. Binary encoding
            4. Target guided encoding
            5. Hash encoding
    6. Imbalance data
            1. collect more data
            2. under sampling 
            3. over sampling
            4. cluster based over sampling
    7. Feature selection
    8. Dimension reduction(PCA, tsne)
    9. Duplicate value treatment
